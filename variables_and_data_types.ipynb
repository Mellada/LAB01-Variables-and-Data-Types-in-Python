# ตัวแปรและชนิดข้อมูลใน Python

ในบทเรียนนี้ เราจะมาทำความเข้าใจเกี่ยวกับตัวแปรและชนิดข้อมูลใน Python ผ่านการทดลองเขียนโค้ดและการเรียนรู้ด้วยตนเอง


## 1. บทนำเกี่ยวกับตัวแปร
ตัวแปร (Variables) ในภาษา Python เป็นสิ่งที่ใช้ในการเก็บข้อมูลหรือค่าต่าง ๆ เพื่อใช้ในโปรแกรม การทำงานของตัวแปรใน Python นั้นค่อนข้างยืดหยุ่นและไม่ต้องกำหนดชนิดของข้อมูลล่วงหน้าเมื่อประกาศตัวแปร เราสามารถใช้ตัวแปรเก็บข้อมูลประเภทต่าง ๆ เช่น ตัวเลข ข้อความ ลิสต์ ฯลฯ ได้ โดย Python จะกำหนดชนิดของตัวแปรอัตโนมัติตามค่าที่เราให้
ตัวอย่างเช่น:


```python
x = 1,2,3,4,5  # ที่นี่ x คือตัวแปรและ 5 คือค่า
print(x)
```

```python
x = "ยินดีต้อนรับ"  # ที่นี่ x คือตัวแปรและ 5 คือค่า # Enclose the Thai text within double quotes to define it as a string.
print(x)
```

## 2. ชนิดข้อมูลใน Python
Python มีชนิดข้อมูลหลายชนิด ที่ถูกใช้ในการเก็บข้อมูลประเภทต่าง ๆ


### a. ตัวเลข (Numbers)
ใน Python ตัวเลขแบ่งเป็นประเภทหลักๆ ดังนี้:

int (จำนวนเต็ม): เก็บค่าจำนวนเต็มทั้งบวกและลบ ไม่มีทศนิยม
float (จำนวนทศนิยม): เก็บค่าตัวเลขที่มีทศนิยม
complex (จำนวนเชิงซ้อน): เก็บค่าจำนวนที่มีทั้งส่วนจริงและส่วนจินตภาพ


```python
x = 10  # จำนวนเต็ม
y = 3.14  # ทศนิยม
z = 1 + 2j  # จำนวนเชิงซ้อน
print(type(x), type(y), type(z))
```

### b. สตริง (Strings)
สตริง (String) ใน Python คือชนิดข้อมูลที่ใช้เก็บลำดับของอักขระ (characters) เช่น ตัวอักษร ตัวเลข สัญลักษณ์ หรือช่องว่าง สตริงถูกใช้บ่อยในการประมวลผลข้อความและข้อมูลที่เป็นข้อความ

การสร้างสตริง
สามารถสร้างสตริงได้โดยใช้เครื่องหมายคำพูดเดี่ยว ('...') หรือเครื่องหมายคำพูดคู่ ("...")


```python
name = "Geographers"
greeting = 'สวัสดี'
print(greeting, name)
```

### c. ลิสต์ (Lists)
ลิสต์ (Lists) ใน Python คือโครงสร้างข้อมูลแบบเรียงลำดับที่สามารถเก็บค่าหลายค่าในตัวแปรเดียวกัน แต่ละค่าภายในลิสต์สามารถเป็นได้ทั้งตัวเลข, สตริง, หรือข้อมูลประเภทอื่นๆ และสามารถเปลี่ยนแปลง (mutable) ได้


```python
fruits = ["apple", "มะละกอ", "กล้วย"]
fruits.append("orange")
print(fruits)
```

### d. ทูเพิล (Tuples)
ทูเพิล (Tuples) ใน Python คือโครงสร้างข้อมูลแบบเรียงลำดับที่เก็บหลายค่าเหมือนลิสต์ แต่มีคุณสมบัติไม่สามารถเปลี่ยนแปลงค่าได้ (immutable) หลังจากสร้างแล้ว

หมายเหตุ: ทูเพิล (Tuple) มีลักษณะคล้ายค่าคงที่ในแง่ที่ว่าข้อมูลภายในไม่สามารถเปลี่ยนแปลงได้หลังจากสร้างแล้ว (immutable) แต่ไม่ใช่ค่าคงที่ในตัวเอง เพราะยังสามารถเก็บข้อมูลหลายค่าในตัวแปรเดียวได้และสามารถถูกใช้ในโค้ดได้ตามปกติ


```python
coordinates = (10, 20)
print(coordinates)
```

### e. ดิกชันนารี (Dictionaries)
ดิกชันนารี (Dictionaries) ใน Python เป็นโครงสร้างข้อมูลแบบไม่เรียงลำดับที่เก็บข้อมูลเป็นคู่ของคีย์ (key) และค่า (value) โดยคีย์จะต้องไม่ซ้ำกัน และใช้คีย์ในการเข้าถึงค่าที่ต้องการ

```python
person = {"name": "John", "age": 30}
print(person["name"], person["age"])
```

### f. บูลีน (Booleans)
เป็นประเภทข้อมูลที่มีค่าเพียงสองค่าเท่านั้น คือ True หรือ False ใช้ในการตรวจสอบเงื่อนไขหรือควบคุมการตัดสินใจในโปรแกรม


```python
is_active = True
is_student = False
print(is_active, is_student)
```

## 3. การตรวจสอบชนิดข้อมูลและการแปลงชนิดข้อมูล


```python
x = 10
print(type(x))  # ตรวจสอบชนิดข้อมูล
y = float(x)  # แปลงเป็นทศนิยม
z = str(x)  # แปลงเป็นสตริง
print(y, z)
```
